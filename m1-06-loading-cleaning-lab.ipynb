```python
import pandas as pd
import numpy as np
```

```python
# --- TASK 1 ---

tickets = pd.read_csv("data_safe_copy.csv")
print(tickets.tail())
print(tickets.head())
print(tickets.info())
```

```python
# --- TASK 2 --- 

tickets["opened_at"] = pd.to_datetime(tickets["opened_at"])
tickets["category"] = tickets["category"].str.lower().str.strip()
tickets["priority"] = tickets["priority"].str.lower().str.strip()
print(tickets["category"].unique())
print(tickets["priority"].unique())
```

```python
# --- TASK 3 ---

tickets.resolution_minutes = pd.to_numeric(tickets['resolution_minutes'], errors='coerce')
print(f"count of missing values : {tickets.resolution_minutes.isna().sum()}")
```

```python
# --- TASK 4 ---

tickets_clean = tickets
print(f" before dropping nan values : {tickets_clean.shape}")
tickets_clean = tickets.dropna(subset=["resolution_minutes"]).copy()
print((tickets_clean["resolution_minutes"] <= 0).sum())
print(f" after dropping nan values : {tickets_clean.shape}")
```

```python
# --- TASK 5 ---

summary = {"total_records" : len(tickets),
           "clean_records" : len(tickets_clean),
           "missing_resolution" : float(tickets.resolution_minutes.isnull().sum()),
           "avg_resolution" : float(tickets.resolution_minutes.mean()),
           "max_resolution" : tickets.resolution_minutes.max()}
summary
```
